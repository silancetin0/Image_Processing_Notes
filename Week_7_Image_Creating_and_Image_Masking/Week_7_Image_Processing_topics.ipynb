```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Renk oluşturma ve maskeleme 
#kod ile kırmızı rengi oluşturma

width,height, channel = 300,300,3

img_red = np.zeros((width,height,channel),dtype=np.uint8)

# kırmızı renk için 
img_red[:,:,0] =255

plt.imshow(img_red)
plt.axis('off')
```

```python
width,height, channel = 300,300,3

img_blue= np.zeros((width,height,channel),dtype=np.uint8)

# kırmızı renk için 
img_blue[:,:,2] =245

plt.imshow(img_blue)
plt.axis('off')
```

```python
##
img_green= np.zeros((width,height,channel),dtype=np.uint8)

# kırmızı renk için 
img_green[:,:,1] =250

plt.imshow(img_green)
plt.axis('off')
```

```python
fig = plt.figure(figsize=(10,7))
rows= 1
columns=3

fig.add_subplot(rows, columns,1)
plt.imshow(img_red)
plt.axis('off')
plt.title("RED Image")

fig.add_subplot(rows, columns,2)
plt.imshow(img_blue)
plt.axis('off')
plt.title('BLUE IMAGE')

fig.add_subplot(rows, columns,3)
plt.imshow(img_green)
plt.axis('off')
plt.title('GREEN IMAGE')

```

```python
# siyah renk cizdirme
img_black= np.zeros((width,height,channel),dtype=np.uint8)
img_white= np.ones((width,height,channel),dtype=np.uint8)*255
img_gray_1= np.zeros((width,height,channel),dtype=np.uint8)

img_gray_1[:,:,:] =128

img_gray_2 = np.random.randint(128,129,(width,height,channel),dtype=np.uint8)
```

```python
fig = plt.figure(figsize=(10,7),facecolor='purple')
rows= 2
columns=2

fig.add_subplot(rows, columns,1)
plt.imshow(img_black)
plt.axis('off')
plt.title("BLACK Image")

fig.add_subplot(rows, columns,2)
plt.imshow(img_white)
plt.axis('off')
plt.title('WHİTE IMAGE')

fig.add_subplot(rows, columns,3)
plt.imshow(img_gray_1)
plt.axis('off')
plt.title('Gray IMAGE')

fig.add_subplot(rows, columns,4)
plt.imshow(img_gray_2)
plt.axis('off')
plt.title('Gray_2 IMAGE')
```

```python
width,height,channel = 300,420,3
img_flag_germany= np.zeros((width,height,channel),dtype=np.uint8)

#0:100 demek ilk 100 satır demek 
img_flag_germany[:100,0]=[0,0,0] #rgb = [0,0,0]

#100:200 demek ilk 100 satırdan 200. satıra kadar
img_flag_germany[100:200,1]=[255,0,0] #rgb = [255,0,0]

#200:300 demek ilk 200 satırdan 300 satıra kadar 
img_flag_germany[200:300,0]=[255,255,0] #rgb = [255,255,0]
```

```python
width, height, channel =300, 420, 3

img_flag_russia = np.zeros((width,height,channel),dtype=np.uint8)

# 0:100 demek ilk 100 satır demektir
img_flag_russia[:100,:,] = [255,255,255] 

# 100:200 demek 100.satırdan 200.satıra kadar anlamına gelir
img_flag_russia[100:200,:,] = [0,0,160]

# 200:300 demek 200.satırdan 300.satıra kadar anlamına gelir
img_flag_russia[200:300,:,] = [255,0,0]    

plt.imshow(img_flag_russia)
plt.axis('off')
plt.title("russia flag")
```

```python
width, height, channel =300, 420, 3

img_flag_belgium = np.zeros((width,height,channel),dtype=np.uint8)

# 0:140 demek ilk 140 sütun demektir
img_flag_belgium[:,:140] = [0,0,0] 

# 140:280 demek 140.sütundan 280.sütuna kadar anlamına gelir
img_flag_belgium[:,140:280] = [255,255,0]

# 280:420 demek 280.sütundan 420.sütuna kadar anlamına gelir
img_flag_belgium[:,280:420] = [255,0,0]    

plt.imshow(img_flag_belgium)
plt.axis('off')
plt.title("belgium flag")
```

```python
mask_red =(img_flag_russia[:, :, 0] == 255) & (img_flag_russia[:, :, 1] == 0 ) & (img_flag_russia[:, :, 2] == 0)

img_flag_russia[mask_red] = [255,255,0]

plt.imshow(img_flag_russia)
plt.axis('off')
                                
```

```python
mask_red =(img_flag_belgium[:, :, 0] == 0) & (img_flag_belgium[:, :, 1] == 0 ) & (img_flag_belgium[:, :, 2] == 0)

img_flag_belgium[mask_red] = [0,0,255]

plt.imshow(img_flag_belgium)
plt.axis('off')
```
